```python
# four square method

# top left box = income
# duplex purchase price: 200,000.00
# rental income 1% of purchase price = duplex, 1,000.00 per each, total 2,000.00 per month
# laundry income = 0
# storage income = 0
# misc income = 0

# bottom left box = expenses, with the two properties together = total 1,610.00
# taxes: 150.00 month
# insurance: 100.00 month
# utilities (electric, water, sewer, garbage, gas): 0
# HOA: 0
# lawn/snow: 0
# vacancy: 5% or rental income, 100.00 month
# repairs (small repairs, holes in the wall): 100.00 month
# capex (capital expenditures (replacements: roof, water heater, appliances, carpet)): 100.00 month
# property management: 10% or rent, 200.00 month
# mortgage: 160,000.00 5% interest over 30 year period, 860.00 month

# top right = cash flow
# income - expenses so 2,000.00 - 1,610.00 = 390.00 monthly cash flow

# bottom right = cash on cash ROI (return on investment), all the below added together is the total investment = 50,000.00
# down payment = 40,000.00
# closing costs = 3,000.00
# rehab budget = 7,000.00
# misc other = 0

# take monthly cash flow, multiply by 12 to get annual cash flow = 4,680.00
# annual cash flow divided by total investment so 4,680.00 / 50,000.00 = 0.0936 = 9.36% is the cash on cash ROI


class ROICalculator():
    
    def __init__(self, income, expenses, cash_flow, investment, cash_ROI):
        self.income = income
        self.expenses = expenses
        self.cash_flow = cash_flow
        self.cash_ROI = cash_ROI
        
    def calc_income(self, income):
#         prompt_house = input("What was the price of the house? ")
        prompt_rental = input("What is the rental price per month? ")
        prompt_laundry = input("What is the laundry income per month? ")
        prompt_storage = input("What is the storage income per month? ")
        prompt_misc = input("What is the total of any other misc rent per month? ")
        self.income = int(prompt_rental) + int(prompt_laundry) + int(prompt_storage) + int(prompt_misc)
        print(f"The total monthly income is: {self.income}")
        
    def calc_expenses(self, expenses):
        prompt_taxes = input("What are the monthly taxes? ")
        prompt_ins = input("What is the monthly insurance? ")
        prompt_utilities = input("What are the monthly utilities? ")
        prompt_HOA = input("What is the monthly HOA? ")
        prompt_lawn = input("What is the monthly lawn maintenance? ")
        prompt_vacancy = input("What is the monthly vacancy? ")
        prompt_repairs = input("What are the monthly repairs? ")
        prompt_capex = input("What are the monthly capital expenditures? ")
        prompt_pm = input("What are the monthly property management fees? ")
        prompt_mortgage = input("What is the monthly mortgage? ")
        self.expenses = int(prompt_taxes) + int(prompt_ins) + int(prompt_utilities) + int(prompt_HOA) + int(prompt_lawn) + int(prompt_vacancy) + int(prompt_repairs) + int(prompt_capex) + int(prompt_pm) + int(prompt_mortgage)
        print(f"The total monthly expense is: {self.expenses}")
        
    def calc_cash_flow(self, cash_flow):
        self.cash_flow = self.income - self.expenses
        print(f"The monthly cash flow is: {self.cash_flow}")
        
    def calc_investment(self, investment):
        prompt_down = input("What was your downpayment? ")
        prompt_closing = input("What were your closing costs? ")
        prompt_rehab = input("What was your rehab cost? ")
        prompt_misc_costs = input("What was your total misc costs? ")
        self.investment = int(prompt_down) + int(prompt_closing) + int(prompt_rehab) + int(prompt_misc_costs)
        print(f"The investment costs are: {self.investment}")
        
    def calc_cash_ROI(self, cash_ROI):
        self.cash_ROI = (self.cash_flow * 12) / self.investment
        print(f"The cash on cash return on investment is: {self.cash_ROI} or "" {:.0%}".format(self.cash_ROI))

user_calc = ROICalculator("", "", "", "", "")

def run():
    while True:
        prompt = input("Would you like to calculate the ROI? yes or no ")
        if prompt == "yes":
            print("Do not use any letters, decimals, or commas... whole numbers only... ")
            user_calc.calc_income("")
            user_calc.calc_expenses("")
            user_calc.calc_cash_flow("")
            user_calc.calc_investment("")
            user_calc.calc_cash_ROI("")
        else:
            break
            
run()
```
